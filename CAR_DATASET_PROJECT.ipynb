```python
total_sales = df["Price ($)"].sum()
avg_price = df["Price ($)"].mean()
cars_sold = df.shape[0]

print("Total Sales:", total_sales)
print("Average Price:", avg_price)
print("Cars Sold:", cars_sold)
```

```python
df.columns = df.columns.str.strip()


df["Date"] = pd.to_datetime(df["Date"])
df["Week"] = df["Date"].dt.isocalendar().week


df["Price ($)"] = pd.to_numeric(df["Price ($)"], errors='coerce')


weekly_sales = df.groupby("Week")["Price ($)"].sum().reset_index()


fig1 = px.line(weekly_sales, x="Week", y="Price ($)",
               title="YTD Sales Weekly Trend")
fig1.show()
```

```python
body_sales = df.groupby("Body Style")["Price ($)"].sum().reset_index()

fig2 = px.pie(body_sales,
              names="Body Style",
              values="Price ($)",
              hole=0.5,
              title="Sales by Body Style")
fig2.show()
```

```python
color_sales = df.groupby("Color")["Price ($)"].sum().reset_index()

fig3 = px.pie(color_sales,
              names="Color",
              values="Price ($)",
              hole=0.5,
              title="Sales by Color")
fig3.show()
```

```python
company_sales = df.groupby("Company").agg({
    "Price ($)": "sum",
    "Model": "count"
}).reset_index()

company_sales.columns = ["Company", "Total Sales", "Cars Sold"]

fig5 = px.bar(company_sales,
              x="Company",
              y="Total Sales",
              title="Company Wise Sales")
fig5.show()
```
